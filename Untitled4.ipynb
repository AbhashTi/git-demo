```python
from gensim.models import Word2Vec
from numpy.linalg import norm
import numpy as np

# Small corpus
sentences = [
    ["i", "like", "ice"],
    ["i", "like", "steam"],
    ["ice", "is", "solid"],
    ["steam", "is", "gas"]
]

# Train model 1
model1 = Word2Vec(
    sentences,
    vector_size=5,
    window=2,
    min_count=1,
    workers=1,
    seed=42
)

# Train model 2 (same data, different seed)
model2 = Word2Vec(
    sentences,
    vector_size=5,
    window=2,
    min_count=1,
    workers=1,
    seed=44
)

# Get vectors
v1_ice = model1.wv["ice"]
v2_ice = model2.wv["ice"]

print("Vector from model 1 (ice):", v1_ice)
print("Vector from model 2 (ice):", v2_ice)

# Cosine similarity function
def cosine_sim(a, b):
    return np.dot(a, b) / (norm(a) * norm(b))

# Compare relationships
sim1 = cosine_sim(model1.wv["ice"], model1.wv["solid"])
sim2 = cosine_sim(model2.wv["ice"], model2.wv["solid"])

print("\nCosine similarity (ice, solid)")
print("Model 1:", sim1)
print("Model 2:", sim2)

```
